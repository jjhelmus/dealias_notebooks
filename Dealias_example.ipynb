```python
import pyart
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
radar = pyart.io.read('single_folded_sweep.nc')
radar.info('c')
```

```python
vel = radar.fields['velocity']['data']
nyquist = radar.instrument_parameters['nyquist_velocity']['data'][0]
```

```python
print vel.min(), vel.max(), nyquist
```

```python
m = plt.imshow(vel.T, vmin=-nyquist, vmax=nyquist)
plt.colorbar(m)
```

```python
import skimage.restoration
```

```python
skimage.restoration.unwrap_phase?
```

```python
vel_scaled = np.pi * vel / nyquist
print vel_scaled.min(), vel_scaled.max() 
```

```python
vel_scaled_unwrapped = skimage.restoration.unwrap_phase(vel_scaled)
```

```python
m = plt.imshow(vel_scaled_unwrapped.T)
plt.colorbar(m)
```

```python
vel_unwrapped = vel_scaled_unwrapped * nyquist / np.pi
```

```python
m = plt.imshow(vel_unwrapped.T)
plt.colorbar(m)
```

```python
radar.add_field_like('velocity', 'corrected_velocity', vel_unwrapped)
```

```python
radar.fields['corrected_velocity']['standard_name'] = 'Corrected Velocity'
radar.fields['velocity']['standard_name'] = 'Raw Velocity'
```

```python
display = pyart.graph.RadarDisplay(radar)
```

```python
display.plot_ppi('velocity', 0, vmin=-nyquist, vmax=nyquist)
```

```python
display.plot_ppi('corrected_velocity', 0, vmin=-2.5*nyquist, vmax=2.5*nyquist)
```
