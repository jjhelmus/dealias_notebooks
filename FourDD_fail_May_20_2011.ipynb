Why the FourDD algorithm fails to unfold May 20, 2011 CSAPR velocities
======================================================================

In this notebook we will examine why the FourDD [1] algorithm often fails to properly unfold
the radial velocity field from the ARM CSAPR radar.  As an example we will look at an example
volume from May 20, 2011.

[1] C. N. James and R. A Houze Jr, A Real-Time Four-Dimensional Doppler
    Dealising Scheme, Journal of Atmospheric and Oceanic Technology, 2001, 18,
    1674, [online](http://journals.ametsoc.org/doi/abs/10.1175/1520-0426%282001%29018%3C1674%3AARTFDD%3E2.0.CO%3B2)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pyart
import netCDF4
import numpy as np
```

```python
pyart._debug_info()
```

Read in the radar volume and examine it

```python
radar = pyart.io.read('105235.mdv')
radar.info('c')
```

Read in the Sounding data from an interpolated sonde file.

```python
target = netCDF4.num2date(radar.time['data'][0], radar.time['units'])
interp_sounde = netCDF4.Dataset('sgpinterpolatedsondeC1.c1.20110520.000000.cdf')
height, speed, direction = pyart.correct.find_time_in_interp_sonde(interp_sounde, target)
```

Perform the dealiasing using FourDD

```python
dealias_data = pyart.correct.dealias_fourdd(
    radar, sounding_heights = height * 1000.0,
    sounding_wind_speeds = speed, sounding_wind_direction = direction,
    keep_original=False)
radar.add_field('corrected_velocity', dealias_data)
```

Plot the first sweep

```python
display = pyart.graph.RadarDisplay(radar)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
display.plot('velocity', 0, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Raw')
ax = fig.add_subplot(122)
display.plot('corrected_velocity', 0, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Corrected')
```

What is wrong with the wedges at ~5:00 and 10:00?  Expand the velocity limits to see what is going wrong. 

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
display.plot('velocity', 0, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Raw')
ax = fig.add_subplot(122)
display.plot('corrected_velocity', 0, vmin=-40, vmax=40, mask_outside=False, 
             colorbar_label='m/s', title='Corrected')
```

For some reason the algorithm is giving unrealistically large radial velocities in these the wedges.  Since the algorithm unfolds the sweeps from the top sweep to the bottom let us look at the top sweep and see if it contains any strange artifacts. 

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
display.plot('velocity', 16, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Raw')
display.set_limits(xlim=(-15, 15), ylim=(-15,15))
ax = fig.add_subplot(122)
display.plot('corrected_velocity', 16, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Corrected')
display.set_limits(xlim=(-15, 15), ylim=(-15,15))
```

Looks to be two regions where the unfolding given odd results, ~3:00-4:30 and a smaller region around ~10:00.  Perhaps those regions have something suspect in the reflectivity or coherent power?

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
display.plot('reflectivity', 16, vmin=-20, vmax=64, mask_outside=False, 
             colorbar_label='dBz', title='Reflectivity')
display.set_limits(xlim=(-15, 15), ylim=(-15,15))
ax = fig.add_subplot(122)
display.plot('normalized_coherent_power', 16, vmin=0, vmax=1, mask_outside=False, 
             colorbar_label='ncp', title='NCP')
display.set_limits(xlim=(-15, 15), ylim=(-15,15))
```

Nope, those fields look alright.  

When dealiasing a sweep, the FourDD algorithm initially finds gates whose velocities are vertically or temporally continuous and then uses these gates to unfold the entire sweep (initial dealiasing).  For the top sweep there are no vertically continuous sweeps to check against (the algorithm only check above the sweep in question) and we have not provided "previously dealiased radial velocity volume" so we cannot check for temporal continuity.  

For the top-most sweep we are limited to checking against the radial velocities interpolated from the sounding data.  Let create this interpolated volume and add it too our radar.  This is done internally when calling the dealias_fourdd function, but we want to examine the results so we pull it out of the function. 

```python
refl_volume = pyart.correct.dealias._create_rsl_volume(radar, 'reflectivity', 0, 131072.0)

height, speed, direction
sounding_heights = height * 1000.0
hc = np.ascontiguousarray(height * 1000, dtype=np.float32)
sc = np.ascontiguousarray(speed, dtype=np.float32)
dc = np.ascontiguousarray(direction, dtype=np.float32)

success, sound_volume = pyart.correct._fourdd_interface.create_soundvolume(
            refl_volume, hc, sc, dc, 1, 0.05)
```

```python
radar.add_field_like('velocity', 'sounding', sound_volume.get_data())
```

Plotting this sounding derivated radial velocity field for the top most sweep 

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
display.plot('velocity', 16, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Raw')
display.set_limits(xlim=(-15, 15), ylim=(-15,15))
ax = fig.add_subplot(122)
display.plot('sounding', 16, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Corrected')
display.set_limits(xlim=(-15, 15), ylim=(-15,15))
```

Lets find the gates which are marked as "good" during the continuity (initial) dealiasing step of the algorithm.  As described in section 3d of the article these are the gates which can be unfolded so that the velcocity is within 25% of the nyquist velocity.  Using only the gates which don't require any unfolding (there may be additional "good" gates but these gates will be included as given).

```python
nyquist_velocity = radar.instrument_parameters['nyquist_velocity']['data'][0]
print nyquist_velocity
```

```python
out_of_range = np.abs(radar.fields['velocity']['data'] - radar.fields['sounding']['data']) >= 0.25 * nyquist_velocity
velocity_after_cont = radar.fields['velocity']['data'].copy()
ma_vel = np.ma.masked_array(velocity_after_cont, mask=out_of_range)
radar.add_field_like('velocity', 'in_range_velocity', ma_vel)
```

Plotting these "good" in range of the sounding velocities along with the other velocities gives us a figure that helps explains why FourDD gives these segments of poor unfolding 

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(221)
display.plot('velocity', 16, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Raw', colorbar_flag=False)
display.set_limits(xlim=(-15, 15), ylim=(-15,15))

ax = fig.add_subplot(222)
display.plot('sounding', 16, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Sounding', colorbar_flag=False)
display.set_limits(xlim=(-15, 15), ylim=(-15,15))

ax = fig.add_subplot(223)
display.plot('in_range_velocity', 16, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Continuous', colorbar_flag=False)
display.set_limits(xlim=(-15, 15), ylim=(-15,15))

ax = fig.add_subplot(224)
display.plot('corrected_velocity', 16, vmin=-20, vmax=20, mask_outside=False, 
             colorbar_label='m/s', title='Corrected', colorbar_flag=False)
display.set_limits(xlim=(-15, 15), ylim=(-15,15))
plt.tight_layout()
```

Notice that in the continuous panel gates from the two artifact regions are included.  This indicates that these gates are marked as "good" in the initial dealiasing step and the velocities are included in the correct velocities field.  Once marked as "good" there is no mechanism in the algorithm to check that these values are in agreement with neighboring gates, once marked good the velocities are "locked in".  Therefore these regions grow during the spatial dealiasing step and window dealiasing step of the algorithm with the final results being region of un-realistic velocities.

Adjusting the `compthresh` parameter of the algorithm can reduce the number of these initially "good" but actually "bad" gates but cannot remove all of them without breaking the algorithm.  The initial dealiasing step must find some gates which agree with the sounding data for the algorithm to preceed, but this nearly always results in some gates being marked as "good" which are in fact not correctly unfolded.
